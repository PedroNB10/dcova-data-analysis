```python
import subprocess
import sys
import os
import ensurepip

def install_pip():
    """Verifica e instala o pip se não estiver presente."""
    try:
        # Verifica se o pip está instalado
        subprocess.check_call([sys.executable, "-m", "pip", "--version"])
        print("Pip já está instalado.")
    except subprocess.CalledProcessError:
        # Tenta instalar o pip usando ensurepip
        print("Pip não está instalado. Instalando o pip...")
        ensurepip.bootstrap()
        subprocess.check_call([sys.executable, "-m", "pip", "install", "--upgrade", "pip"])
        print("Pip foi instalado com sucesso.")

def install_requirements():
    """Instala pacotes do arquivo requirements.txt."""
    requirements_file = 'requirements.txt'
    
    # Verifica se o arquivo requirements.txt existe
    if os.path.exists(requirements_file):
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", "-r", requirements_file])
            print("Pacotes instalados com sucesso.")
        except subprocess.CalledProcessError as e:
            print(f"Erro ao instalar pacotes: {e}")
    else:
        print(f"Arquivo {requirements_file} não encontrado.")

if __name__ == "__main__":
    install_pip()  # Verifica e instala o pip se necessário
    install_requirements()  # Tenta instalar os pacotes do requirements.txt

import subprocess
import sys
import os
import ensurepip

def install_pip():
    """Verifica e instala o pip se não estiver presente."""
    try:
        # Verifica se o pip está instalado
        subprocess.check_call([sys.executable, "-m", "pip", "--version"])
        print("Pip já está instalado.")
    except subprocess.CalledProcessError:
        # Tenta instalar o pip usando ensurepip
        print("Pip não está instalado. Instalando o pip...")
        ensurepip.bootstrap()
        subprocess.check_call([sys.executable, "-m", "pip", "install", "--upgrade", "pip"])
        print("Pip foi instalado com sucesso.")

def install_requirements():
    """Instala pacotes do arquivo requirements.txt."""
    requirements_file = 'requirements.txt'
    
    # Verifica se o arquivo requirements.txt existe
    if os.path.exists(requirements_file):
        try:
            subprocess.check_call([sys.executable, "-m", "pip", "install", "-r", requirements_file])
            print("Pacotes instalados com sucesso.")
        except subprocess.CalledProcessError as e:
            print(f"Erro ao instalar pacotes: {e}")
    else:
        print(f"Arquivo {requirements_file} não encontrado.")

if __name__ == "__main__":
    install_pip()  # Verifica e instala o pip se necessário
    install_requirements()  # Tenta instalar os pacotes do requirements.txt
```
